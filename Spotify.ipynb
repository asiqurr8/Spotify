```python
#creating a  scatter plot

# Create a DataFrame
df = pd.read_csv('/spotify.csv')

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['Rank'], color='blue', s=100)

# Add titles and labels
plt.title('Scatter Plot of Song Rankings', fontsize=14)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Rank', fontsize=12)

# Annotate each point with the song title and artist
for i, row in df.iterrows():
    plt.text(i, row['Rank'], f'{row["Song Title"]}\n{row["Artist"]}', fontsize=9, ha='center')

# Show plot
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()
```

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(df['Song Title'], df['Rank'], color='lightblue')

# Add title and labels
plt.title('Song Rankings')
plt.xlabel('Song Title')
plt.ylabel('Rank')

# Invert y-axis to show higher ranks (1, 2, 3...) at the top
plt.gca().invert_xaxis()

# Show plot
plt.xticks(rotation=90)  # Rotate song titles for better readability
plt.show()
```

```python
# Matrix for each song
matrix = df.pivot(index='Song Title', columns='Rank', values='Rank')

# Heatmap plot
plt.figure(figsize=(10, 6))
sns.heatmap(matrix, annot=True, cmap='coolwarm', cbar=True, linewidths=1, linecolor='gray')

# Add titles and labels
plt.title('Heatmap of Song Rankings', fontsize=14)
plt.xlabel('Rank', fontsize=12)
plt.ylabel('Song Title', fontsize=12)

# Show plot
plt.show()
```

```python
#Calculate mean, median, and standard deviation
mean_rank = df['Rank'].mean()
median_rank = df['Rank'].median()
std_rank = df['Rank'].std()

print(f"Mean : {mean_rank}")
print(f"Median : {median_rank}")
print(f"Standard Deviation : {std_rank}")
```

```python
#Numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Print the correlation matrix
print(corr_matrix)
```
